```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
data=pd.read_excel(r'C:\Users\ADMIN\Downloads\student_data.xlsx')
data.head()
```

```python
plt.scatter(data['Hours'], data['Scores'], color='blue')
plt.title("Hours vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
X = data[['Hours']]
y = data['Scores']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("R² Score:", r2_score(y_test, y_pred))
print("RMSE:", root_mean_squared_error(y_test, y_pred))
```

```python
line = model.coef_ * X + model.intercept_

plt.scatter(X, y)
plt.plot(X, line, color='red')
plt.title("Regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()

study_hours = 10
predicted_score = model.predict(pd.DataFrame([[study_hours]], columns=['Hours']))
print(f"Predicted score if student studies {study_hours} hours: {predicted_score[0]:.2f}")
```
